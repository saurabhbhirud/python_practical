Python Dictionary Comprehension

```python
dict1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
# Double each value in the dictionary
dict2 = {k: v * 2 for (k, v) in dict1.items()}
print(dict1)
print(dict2)
```

Dictionary comprehension using for loop

```python
# Using for loop
numbers = range(10)
new_dict_for = {}
# Add values to `new_dict_for` using for loop
for n in numbers:
    if n % 2 == 0:
        new_dict_for[n] = n ** 2

print(new_dict_for)
```
