AIM: Program to find Pythagoras triplets.

CODE: Pythagoras triplets

```python
L = int(input("Enter a limit: "))
count = 0

for a in range(1, L):
    for b in range(a + 1, L):
        for c in range(b + 1, L + 1):
            x = a * a
            y = b * b
            z = c * c
            if x + y == z:
                count += 1
                print(a, b, c)

print(f"Total number of Pythagorean triples found: {count}")
```
