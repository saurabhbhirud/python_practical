CODE: Open a file and read the contents of a file and print

```python
read_file = open("readme_test.txt", "r")
print(read_file.read())
read_file.close()
```

```python
# Open the file in append mode and add the text with a newline character
testfile = open("readme_test.txt", "a")
testfile.write("\nYou are the most valuable asset of the Nation.\n")
testfile.close()

# Read the updated file and print its contents
appended_file = open("readme_test.txt", "r")
print(appended_file.read())
appended_file.close()
```

CODE: Open a file and write to a file â€“ Overwrite mod

```python
# Open the file in write mode and write a line
with open("readme_test.txt", "w") as testfile:
    testfile.write("You are most valuable assets of Nation \n")

# Open the file in read mode and print its contents
with open("readme_test.txt", "r") as appended_file:
    print(appended_file.read())
```

CODE: To count number of lines, words and character in file

```python
file = open("readme_test.txt", "r")

no_of_lines = 0
no_of_words = 0
no_of_characters = 0

for line in file:
    line = line.strip("\n")

    words = line.split()

    no_of_lines += 1
    no_of_words += len(words)
    no_of_characters += len(line)

file.close()

# Print the results
print("Lines:", no_of_lines, "Words:", no_of_words, "Characters:", no_of_characters)
```
